```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
timefile = open("C:/bla/Waste/MA/Data/check.log", "r")
file = timefile.readlines()

processedFiles = []
usedSeconds = []

for line in file:
    if line.startswith("Verarbeitet"): 
        newLines = line.split()
        processedFiles.append(newLines[1])
        usedSeconds.append(float(newLines[6]))
```

```python
newListUsedSeconds = []

for i in range(len(usedSeconds) - 1):
    j = i + 1
    newListUsedSeconds.append(usedSeconds[j] - usedSeconds[i])

print(newListUsedSeconds)
```

```python
np.mean(newListUsedSeconds)
```

```python
def plot_scatter(x, y, name):

        # Plot erstellen
        plt.figure(figsize=(14, 8))
        plt.plot(x, y)
        plt.scatter(x, y)

        # Achsentitel und Beschriftungen
        plt.title(name, fontsize=16)
        plt.xlabel("Process for every 1000 samples", fontsize=14)
        plt.ylabel("Time (seconds)", fontsize=14)

        # Gitternetz hinzufügen
        plt.grid(True, linestyle='--', alpha=0.7)

        # Legende anzeigen
        plt.legend(fontsize=12)

        # Plot anzeigen
        plt.tight_layout()
        plt.show()
```

```python
plot_scatter(processedFiles[0:15], newListUsedSeconds[0:15], "RNAz")
```
